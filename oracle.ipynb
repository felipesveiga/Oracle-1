```python
# Crreating a class to proceed DML statements
class DML:
    def __init__(self, table_name, query, params = None):
        self.table_name = table_name
        self.query = query
        self.params = params
    
    def command(self, query, values = None, params = None):
        for value, parameter in zip(values, params):
            cx_Oracle.execute(self.query, values)
```

```python
import matplotlib.pyplot as plt
import cx_Oracle
dsn = cx_Oracle.makedsn('localhost.localdomain', 1521, service_name = 'orcl')
connection = cx_Oracle.connect('OT', 'oracle', dsn = dsn)
query = ''' SELECT O.ORDER_DATE, PC.CATEGORY_NAME, SUM(OI.UNIT_PRICE * OI.QUANTITY) REVENUE
            FROM ORDER_ITEMS OI, ORDERS O, PRODUCTS P, PRODUCT_CATEGORIES PC
            WHERE PC.CATEGORY_ID = P.CATEGORY_ID
            AND P.PRODUCT_ID = OI.PRODUCT_ID
            AND O.ORDER_ID = OI.ORDER_ID
            GROUP BY PC.CATEGORY_NAME, O.ORDER_DATE
            ORDER BY O.ORDER_DATE
'''


a = pd.read_sql(query, connection,parse_dates = True, index_col = ['ORDER_DATE'])
a.index = a.index.to_period('Y')
a = a.groupby([a.index, 'CATEGORY_NAME']).sum()
products = [product for product in a.index.get_level_values(1).unique()]
l = [(i, j) for i in range(2) for j in range(2)]

fig, ax = plt.subplots(nrows = 2, ncols = 2, sharex= False, sharey = False, figsize = (22,8)) 

for product, position in zip(products,l):
    x = a.xs(key= product, level=1)
    x.plot(ax = ax[position] )
    ax[position].set_title(product)
    ax[position].set_ylabel('REVENUE (MILLIONS $)')

plt.tight_layout()
#ax[1,1].set_title('Mother Board')
```

```python
import matplotlib.pyplot as plt
import cx_Oracle
dsn = cx_Oracle.makedsn('localhost.localdomain', 1521, service_name = 'orcl')
connection = cx_Oracle.connect('OT', 'oracle', dsn = dsn)
query = ''' SELECT O.ORDER_DATE, PC.CATEGORY_NAME, SUM(OI.UNIT_PRICE * OI.QUANTITY) REVENUE
            FROM ORDER_ITEMS OI, ORDERS O, PRODUCTS P, PRODUCT_CATEGORIES PC
            WHERE PC.CATEGORY_ID = P.CATEGORY_ID
            AND P.PRODUCT_ID = OI.PRODUCT_ID
            AND O.ORDER_ID = OI.ORDER_ID
            GROUP BY PC.CATEGORY_NAME, O.ORDER_DATE
            ORDER BY O.ORDER_DATE
'''


a = pd.read_sql(query, connection,parse_dates = True, index_col = ['ORDER_DATE'])
a.index = a.index.to_period('Y')
a = a.groupby([a.index, 'CATEGORY_NAME']).sum()
products = [product for product in a.index.get_level_values(1).unique()]
l = [(i, j) for i in range(2) for j in range(2)]

fig, ax = plt.subplots(figsize = (22,8)) 

cpu = a.xs(key = 'CPU', level = 1)#.plot(kind = 'bar', width = 0.3, ax = ax)
display(cpu)
x = [i for i in range (len(cpu))]
print(x)
for i, value in zip(x, cpu.REVENUE):
    plt.bar(i+ 0.3, value, color = 'b', width = 0.3)
    
storage = a.xs(key = 'Storage', level = 1).plot(kind = 'bar', color = 'orange', width = 0.3, ax = ax)
for i, value in zip(x, storage.REVENUE):
    plt.bar(i, value, color = 'orange', width= 0.3)

#plt.tight_layout()
#ax[1,1].set_title('Mother Board')
```
